# Fixtures

```python
home_team = []
score = []
away_team = []
referees = []
attendances = []
matches = []
match_links = []
match_report_links = []
#dates = []

for i in link_list:
    print(i)
    response = requests.get(i)
    soup = BeautifulSoup(response.content, "html.parser")    

#url = 'https://fbref.com/en/comps/24/schedule/Serie-A-Scores-and-Fixtures'
#response = requests.get(url)
#soup = BeautifulSoup(response.content, "html.parser")

    for match in soup.find_all('table', class_='stats_table'):
        #date = match.find_all('td', {"class": "left", 'data-stat': 'date'})
        home = match.find_all('td', {"class": "right", 'data-stat': 'squad_a'})
        result = match.find_all('td', {"class": "center", 'data-stat': 'score'})
        away = match.find_all('td', {"class": "left", 'data-stat': 'squad_b'})
        attendance = match.find_all('td', {"class": "right", 'data-stat': 'attendance'})
        referee = match.find_all('td', {"class": "left", 'data-stat': 'referee'})
        match_report = match.find_all('a')

        for i in home:
            home_team.append(i.text)

        for i in result:
            score.append(i.text)

        for i in away:
            away_team.append(i.text)
            
        for i in referee:
            referees.append(i.text)
            
        for i in attendance:
            attendances.append(i.text)
            
        for link in match_report:
            i = str(link)
            match_links.append(i)
            
        #for i in date:
         #   dates.append(i.string)
            
            
# Adjust and select the correct match_report links.            
for i in match_links:
    a = i.strip('</a>')
    b = a.strip(' href="')
    if b.endswith('Match Report'):
        match_report_links.append('https://fbref.com' + b)
```

```python
def stripping_home(x):
    x = x.str.strip('[(').str.strip('])')
    x = x.str.split(',', 1, expand=True)
    x[0] = x[0].str.strip("'")
    return x[0]
```

```python
def stripping_away(x):
    x = x.str.strip('[(').str.strip('])')
    x = x.str.split("'", 5, expand=True)
    x[5] = x[5].str.strip("'")
    return x[5]
```

```python
df['HomeFouls'] = stripping_home(df.HomeFouls)
df['HomeCorners'] = stripping_home(df.HomeCorners)
df['HomeCrosses'] = stripping_home(df.HomeCrosses)
df['HomeTouches'] = stripping_home(df.HomeTouches)
df['HomeTackles'] = stripping_home(df.HomeTackles)
df['HomeInterceptions'] = stripping_home(df.HomeInterceptions)
df['HomeAerials'] = stripping_home(df.HomeAerials)
df['HomeClearances'] = stripping_home(df.HomeClearances)
df['HomeOffsides'] = stripping_home(df.HomeOffsides)
df['HomeGoalKicks'] = stripping_home(df.HomeGoalKicks)
df['HomeThrowIns'] = stripping_home(df.HomeThrowIns)
df['HomeLongBalls'] = stripping_home(df.HomeLongBalls)
```

```python
df['AwayFouls'] = stripping_away(df.AwayFouls)
df['AwayCorners'] = stripping_away(df.AwayCorners)
df['AwayCrosses'] = stripping_away(df.AwayCrosses)
df['AwayTouches'] = stripping_away(df.AwayTouches)
df['AwayTackles'] = stripping_away(df.AwayTackles)
df['AwayInterceptions'] = stripping_away(df.AwayInterceptions)
df['AwayAerials'] = stripping_away(df.AwayAerials)
df['AwayClearances'] = stripping_away(df.AwayClearances)
df['AwayOffsides'] = stripping_away(df.AwayOffsides)
df['AwayGoalKicks'] = stripping_away(df.AwayGoalKicks)
df['AwayThrowIns'] = stripping_away(df.AwayThrowIns)
df['AwayLongBalls'] = stripping_away(df.AwayLongBalls)
```
