# Fixtures

```python
home_team = []
score = []
away_team = []
referees = []
attendances = []
matches = []
match_links = []
match_report_links = []

for i in link_list:
    print(i)
    response = requests.get(i)
    soup = BeautifulSoup(response.content, "html.parser")    

#url = 'https://fbref.com/en/comps/24/schedule/Serie-A-Scores-and-Fixtures'
#response = requests.get(url)
#soup = BeautifulSoup(response.content, "html.parser")

    for match in soup.find_all('table', class_='stats_table'):
        home = match.find_all('td', {"class": "right", 'data-stat': 'squad_a'})
        result = match.find_all('td', {"class": "center", 'data-stat': 'score'})
        away = match.find_all('td', {"class": "left", 'data-stat': 'squad_b'})
        attendance = match.find_all('td', {"class": "right", 'data-stat': 'attendance'})
        referee = match.find_all('td', {"class": "left", 'data-stat': 'referee'})
        match_report = match.find_all('a')

        for i in home:
            home_team.append(i.text)

        for i in result:
            score.append(i.text)

        for i in away:
            away_team.append(i.text)
            
        for i in referee:
            referees.append(i.text)
            
        for i in attendance:
            attendances.append(i.text)
            
        for link in match_report:
            i = str(link)
            match_links.append(i)
            
            
# Adjust and select the correct match_report links.            
for i in match_links:
    a = i.strip('</a>')
    b = a.strip(' href="')
    if b.endswith('Match Report'):
        match_report_links.append('https://fbref.com' + b)
```

## Transform it in a DataFrame

```python
df_soccer = pd.DataFrame(a[:, 0], columns=['match_report_link', 'fouls',
                             'corners', 'crosses', 'touches',
                             'tackles', 'interceptions', 'aerials_won',
                             'clearances', 'offsides', 'goal_kicks',
                             'throw_ins', 'long_balls'])
df_soccer = df_soccer.rename(columns={'fouls': 'HomeFouls', 'corners': 'HomeCorners', 'crosses': 'HomeCrosses', 'touches':'HomeTouches', 
                  'tackles': 'HomeTackles', 'interceptions': 'HomeInterceptions', 'aerials_won': 'HomeAerials', 'clearances':'HomeClearances',
                 'offsides': 'HomeOffsides', 'goal_kicks': 'HomeGoalKicks', 'throw_ins': 'HomeThrowIns', 'long_balls': 'HomeLongBalls'})
df_soccer_copy = df_soccer.copy()
df_soccer_copy = df_soccer_copy.rename(columns={'HomeFouls': 'AwayFouls', 'HomeCorners': 'AwayCorners', 'HomeCrosses': 'AwayCrosses', 'HomeTouches':'AwayTouches', 
                  'HomeTackles': 'AwayTackles', 'HomeInterceptions': 'AwayInterceptions', 'HomeAerials': 'AwayAerials', 'HomeClearances':'AwayClearances',
                 'HomeOffsides': 'AwayOffsides', 'HomeGoalKicks': 'AwayGoalKicks', 'HomeThrowIns': 'AwayThrowIns', 'HomeLongBalls': 'AwayLongBalls', 'match_report_link': 'match_report_link1'})
df_soccer_full = pd.concat([df_soccer, df_soccer_copy], axis=1)
df_soccer_full = df_soccer_full.drop(columns='match_report_link1')
```

```python
#extract values
def grab_first(x):
  return x[0][0] if x else -1

def grab_last(x):
  return x[0][-1] if x else -1
```

```python
  links_df = df_soccer_full['match_report_link']
  home_df = df_soccer_full.filter(regex='Home', axis=1).applymap(grab_first)
  away_df = df_soccer_full.filter(regex='Away', axis=1).applymap(grab_last)
```

```python
df = pd.concat((links_df, home_df, away_df), axis=1)

df.head(5) 
```

```python
matches_df["Home_Poss"] = matches_df["Home_Poss"].str.rstrip("%")
matches_df["Home_Poss"] = pd.to_numeric(matches_df["Home_Poss"])
matches_df["Home_Poss"] = matches_df["Home_Poss"] / 100.0
matches_df["Away_Poss"] = matches_df["Away_Poss"].str.rstrip("%")
matches_df["Away_Poss"] = pd.to_numeric(matches_df["Away_Poss"])
matches_df["Away_Poss"] = matches_df["Away_Poss"] / 100.0

matches_df["attendance"] = matches_df["attendance"].str.replace(",", "")
matches_df["attendance"] = pd.to_numeric(matches_df["attendance"])

matches_df["Home_PA"] = matches_df["Home_PA"].str.rstrip("%")
matches_df["Away_PA"] = matches_df["Away_PA"].str.rstrip("%")
matches_df["Home_PA"] = pd.to_numeric(matches_df["Home_PA"])
matches_df["Away_PA"] = pd.to_numeric(matches_df["Away_PA"])

matches_df["Home_ShoT"] = matches_df["Home_ShoT"].str.rstrip("%")
matches_df["Away_ShoT"] = matches_df["Away_ShoT"].str.rstrip("%")
matches_df["Home_ShoT"] = pd.to_numeric(matches_df["Home_ShoT"])
matches_df["Away_ShoT"] = pd.to_numeric(matches_df["Away_ShoT"])

matches_df["Home_Saves"] = matches_df["Home_Saves"].str.rstrip("%")
matches_df["Away_Saves"] = matches_df["Away_Saves"].str.rstrip("%")
matches_df["Home_Saves"] = pd.to_numeric(matches_df["Home_Saves"])
matches_df["Away_Saves"] = pd.to_numeric(matches_df["Away_Saves"])

matches_df["HomeGoals"] = pd.to_numeric(matches_df["HomeGoals"])
matches_df["AwayGoals"] = pd.to_numeric(matches_df["AwayGoals"])
```
