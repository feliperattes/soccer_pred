```python
matches_df["Home_Poss"] = matches_df["Home_Poss"].str.rstrip("%")
matches_df["Home_Poss"] = pd.to_numeric(matches_df["Home_Poss"])
matches_df["Home_Poss"] = matches_df["Home_Poss"] / 100.0

matches_df["Away_Poss"] = matches_df["Away_Poss"].str.rstrip("%")
matches_df["Away_Poss"] = pd.to_numeric(matches_df["Away_Poss"])
matches_df["Away_Poss"] = matches_df["Away_Poss"] / 100.0
```

```python
matches_df["attendance"] = matches_df["attendance"].str.replace(",", "")
matches_df["attendance"] = pd.to_numeric(matches_df["attendance"])
```

```python
matches_df["Home_PA"] = matches_df["Home_PA"].str.rstrip("%")
matches_df["Away_PA"] = matches_df["Away_PA"].str.rstrip("%")

matches_df["Home_PA"] = pd.to_numeric(matches_df["Home_PA"])
matches_df["Away_PA"] = pd.to_numeric(matches_df["Away_PA"])
```

```python
matches_df["Home_ShoT"] = matches_df["Home_ShoT"].str.rstrip("%")
matches_df["Away_ShoT"] = matches_df["Away_ShoT"].str.rstrip("%")

matches_df["Home_ShoT"] = pd.to_numeric(matches_df["Home_ShoT"])
matches_df["Away_ShoT"] = pd.to_numeric(matches_df["Away_ShoT"])
```

```python
matches_df["Home_Saves"] = matches_df["Home_Saves"].str.rstrip("%")
matches_df["Away_Saves"] = matches_df["Away_Saves"].str.rstrip("%")

matches_df["Home_Saves"] = pd.to_numeric(matches_df["Home_Saves"])
matches_df["Away_Saves"] = pd.to_numeric(matches_df["Away_Saves"])
```

```python
for team in teams:
    matches_df.loc[matches_df["HT"] == team, "Total Home Matches"] = matches_df[(matches_df["HT"] == team)].count()[0] 
    matches_df.loc[matches_df["AT"] == team, "Total Away Matches"] = matches_df[(matches_df["AT"] == team)].count()[0] 
```

```python
matches_df[matches_df["HT"] == "Santos"]
```

```python
for team in teams:
    #comulative
    #matches_df.loc[matches_df["Home"] == team, "Goals Sum HOME"] = matches_df[(matches_df["Home"] == team)]["HG"].cumsum()
    #matches_df.loc[matches_df["Away"] == team, "Goals Sum AWAY"] = matches_df[(matches_df["Away"] == team)]["AG"].cumsum()
    #total Corners
    matches_df.loc[matches_df["HT"] == team, "HomeCornersSum"] = matches_df[(matches_df["HT"] == team)]["HomeCorners"].sum()
    matches_df.loc[matches_df["AT"] == team, "AwayCornersSum"] = matches_df[(matches_df["AT"] == team)]["AwayCorners"].sum()
    #total Crosses
    matches_df.loc[matches_df["HT"] == team, "HomeCrossesSum"] = matches_df[(matches_df["HT"] == team)]["HomeCrosses"].sum()
    matches_df.loc[matches_df["AT"] == team, "AwayCrossesSum"] = matches_df[(matches_df["AT"] == team)]["AwayCrosses"].sum()
    #total Poss
    matches_df.loc[matches_df["HT"] == team, "HomePossSum"] = matches_df[(matches_df["HT"] == team)]["Home_Poss"].sum()
    matches_df.loc[matches_df["AT"] == team, "AwayPossSum"] = matches_df[(matches_df["AT"] == team)]["Away_Poss"].sum()
    #total Throws
    matches_df.loc[matches_df["HT"] == team, "HomeThrowsSum"] = matches_df[(matches_df["HT"] == team)]["HomeThrowIns"].sum()
    matches_df.loc[matches_df["AT"] == team, "AwayThrowsSum"] = matches_df[(matches_df["AT"] == team)]["AwayThrowIns"].sum()
```

```python
matches_df["Average_HomeTeam_Corners"] = matches_df["HomeCornersSum"] / matches_df["Total Home Matches"]
matches_df["Average_AwayTeam_Corners"] = matches_df["AwayCornersSum"] / matches_df["Total Away Matches"]

matches_df["Average_HomeTeam_Crosses"] = matches_df["HomeCrossesSum"] / matches_df["Total Home Matches"]
matches_df["Average_AwayTeam_Crosses"] = matches_df["AwayCrossesSum"] / matches_df["Total Away Matches"]

matches_df["Average_HomeTeam_Poss"] = matches_df["HomePossSum"] / matches_df["Total Home Matches"]
matches_df["Average_AwayTeam_Poss"] = matches_df["AwayPossSum"] / matches_df["Total Away Matches"]

matches_df["Average_HomeTeam_Throws"] = matches_df["HomeThrowsSum"] / matches_df["Total Home Matches"]
matches_df["Average_AwayTeam_Throws"] = matches_df["AwayThrowsSum"] / matches_df["Total Away Matches"]
```
