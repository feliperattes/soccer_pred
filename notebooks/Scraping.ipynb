# Fixtures

```python
home_team = []
score = []
away_team = []
referees = []
attendances = []
matches = []
match_links = []
match_report_links = []

for i in link_list:
    print(i)
    response = requests.get(i)
    soup = BeautifulSoup(response.content, "html.parser")    

#url = 'https://fbref.com/en/comps/24/schedule/Serie-A-Scores-and-Fixtures'
#response = requests.get(url)
#soup = BeautifulSoup(response.content, "html.parser")

    for match in soup.find_all('table', class_='stats_table'):
        home = match.find_all('td', {"class": "right", 'data-stat': 'squad_a'})
        result = match.find_all('td', {"class": "center", 'data-stat': 'score'})
        away = match.find_all('td', {"class": "left", 'data-stat': 'squad_b'})
        attendance = match.find_all('td', {"class": "right", 'data-stat': 'attendance'})
        referee = match.find_all('td', {"class": "left", 'data-stat': 'referee'})
        match_report = match.find_all('a')

        for i in home:
            home_team.append(i.text)

        for i in result:
            score.append(i.text)

        for i in away:
            away_team.append(i.text)
            
        for i in referee:
            referees.append(i.text)
            
        for i in attendance:
            attendances.append(i.text)
            
        for link in match_report:
            i = str(link)
            match_links.append(i)
            
            
# Adjust and select the correct match_report links.            
for i in match_links:
    a = i.strip('</a>')
    b = a.strip(' href="')
    if b.endswith('Match Report'):
        match_report_links.append('https://fbref.com' + b)
```
