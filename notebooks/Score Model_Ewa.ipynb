```python
target = ["result"]
```

```python
X = matches_df.drop(target, axis="columns") 
y = matches_df[target]

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=.20,random_state=7)
```

```python
model = LogisticRegression()

model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```
