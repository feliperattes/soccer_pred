```python
import pandas as pd
import numpy as np
from sklearn.linear_model import Ridge
from sklearn.preprocessing import MinMaxScaler
```

```python
matches_df = pd.read_csv("soccer_final.csv")
```

```python
matches_df
```

```python
matches_df = matches_df.drop(["Unnamed: 0"], axis=1)
matches_df
```

```python
matches_df['goal_difference'] = matches_df['HomeGoals'] - matches_df['AwayGoals']
```

```python
matches_df['home_win'] = np.where(matches_df['goal_difference'] > 0, 1, 0)
matches_df['home_loss'] = np.where(matches_df['goal_difference'] < 0, 1, 0)
```

```python
matches_df.columns
```

```python
matches_df = matches_df[["HT","HomeGoals", "AT", "AwayGoals", "Score", "goal_difference", 'home_win', 'home_loss', 'attendance', 'referee', 'Home_Poss', 'Away_Poss', 'Home_PA', 'Away_PA',
       'Home_ShoT', 'Away_ShoT', 'Home_Saves', 'Away_Saves', 'HomeFouls',
       'HomeCorners', 'HomeCrosses', 'HomeTouches', 'HomeTackles',
       'HomeInterceptions', 'HomeAerials', 'HomeClearances', 'HomeOffsides',
       'HomeGoalKicks', 'HomeThrowIns', 'HomeLongBalls', 'AwayFouls',
       'AwayCorners', 'AwayCrosses', 'AwayTouches', 'AwayTackles',
       'AwayInterceptions', 'AwayAerials', 'AwayClearances', 'AwayOffsides',
       'AwayGoalKicks', 'AwayThrowIns', 'AwayLongBalls','match_report_link']]
```

```python
matches_df
```

```python
df_visitor = pd.get_dummies(matches_df['AT'], dtype=np.int64)
df_home = pd.get_dummies(matches_df['HT'], dtype=np.int64)
```

```python
df_model = df_home.sub(df_visitor) 
df_model['goal_difference'] = matches_df['goal_difference']
```

```python
df_model
```

```python
df_train = df_model # not required but I like to rename my dataframe with the name train.

lr = Ridge(alpha=0.001) 
X = df_train.drop(['goal_difference'], axis=1)
y = df_train['goal_difference']

lr.fit(X, y)
```

```python
df_ratings = pd.DataFrame(data={'team': X.columns, 'rating': lr.coef_})
df_ratings
```

```python
df_ratings.rating.max()
```

```python
result = pd.concat([matches_df, df_model], axis=1, sort=False)
```

```python
full_report_df = result
```

```python
full_report_df.columns
```

```python
full_report_df["Home_Poss"] = full_report_df["Home_Poss"].str.rstrip("%")
```

```python
full_report_df["Home_Poss"]
```

```python
full_report_df["Home_Poss"] = pd.to_numeric(full_report_df["Home_Poss"])
```

```python
full_report_df["Home_Poss"]
```

```python
full_report_df["Home_Poss"] = full_report_df["Home_Poss"] / 100.0
```

```python
full_report_df["attendance"] = full_report_df["attendance"].str.replace(",", "")
```

```python
full_report_df["attendance"] = pd.to_numeric(full_report_df["attendance"])
```

```python
full_report_df["Away_Poss"] = full_report_df["Away_Poss"].str.rstrip("%")
```

```python
full_report_df["Away_Poss"] = pd.to_numeric(full_report_df["Away_Poss"])
```

```python
full_report_df["Home_PA"] = full_report_df["Home_PA"].str.rstrip("%")
full_report_df["Away_PA"] = full_report_df["Away_PA"].str.rstrip("%")

full_report_df["Home_PA"] = pd.to_numeric(full_report_df["Home_PA"])
full_report_df["Away_PA"] = pd.to_numeric(full_report_df["Away_PA"])

full_report_df.dtypes
```

```python
full_report_df["Home_ShoT"] = full_report_df["Home_ShoT"].str.rstrip("%")
full_report_df["Away_ShoT"] = full_report_df["Away_ShoT"].str.rstrip("%")

full_report_df["Home_ShoT"] = pd.to_numeric(full_report_df["Home_ShoT"])
full_report_df["Away_ShoT"] = pd.to_numeric(full_report_df["Away_ShoT"])

full_report_df.dtypes
```

```python
full_report_df["Home_Saves"] = full_report_df["Home_Saves"].str.rstrip("%")
full_report_df["Away_Saves"] = full_report_df["Away_Saves"].str.rstrip("%")

full_report_df["Home_Saves"] = pd.to_numeric(full_report_df["Home_Saves"])
full_report_df["Away_Saves"] = pd.to_numeric(full_report_df["Away_Saves"])

full_report_df.dtypes
```
