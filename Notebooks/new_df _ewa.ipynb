```python
import pandas as pd
import numpy as np
import re
import string
```

```python
df_soccer = pd.read_csv('soccer.csv')
```

```python
df_soccer = df_soccer.rename(columns={'0': 'HomeFouls', '1': 'HomeCorners', '2': 'HomeCrosses', '3':'HomeTouches', 
                  '4': 'HomeTackles', '5': 'HomeInterceptions', '6': 'HomeAerials', '7':'HomeClearances',
                 '8': 'HomeOffsides', '9': 'HomeGoalKicks', '10': 'HomeThrowIns', '11': 'HomeLongBalls'})
```

```python
df_soccer_copy = df_soccer.copy()
df_soccer_copy = df_soccer_copy.rename(columns={'HomeFouls': 'AwayFouls', 'HomeCorners': 'AwayCorners', 'HomeCrosses': 'AwayCrosses', 'HomeTouches':'AwayTouches', 
                  'HomeTackles': 'AwayTackles', 'HomeInterceptions': 'AwayInterceptions', 'HomeAerials': 'AwayAerials', 'HomeClearances':'AwayClearances',
                 'HomeOffsides': 'AwayOffsides', 'HomeGoalKicks': 'AwayGoalKicks', 'HomeThrowIns': 'AwayThrowIns', 'HomeLongBalls': 'AwayLongBalls'})
```

```python
# dataset with all the columns we want, now we need to extract the values
df_soccer_full = pd.concat([df_soccer, df_soccer_copy], axis=1)
```

```python
df_soccer_full.head(2)
```

```python
#for loop with regex

columns = df_soccer_full.columns
columns

for i in columns:
    df_soccer_full[f'{i}'] = df_soccer_full[f'{i}'].apply(lambda x: re.findall(r'\d+', x))

#df_soccer_full['HomeFouls'] = df_soccer_full.HomeFouls.apply(lambda x: re.findall(r'\d+', x))
```

```python
df_soccer_full.head(2)
```

```python
# grabbing the first item of the list  - all cols

home_col = df_soccer_full.columns[0:12]
away_col = df_soccer_full.columns[12:24]

for i in home_col:
    df_soccer_full[f'{i}'] = df_soccer_full[f'{i}'].apply(lambda x: x[0] if len(x) == 2 else 'NaN')
    
for col in away_col:
    df_soccer_full[f'{col}'] = df_soccer_full[f'{col}'].apply(lambda x: x[1] if len(x) == 2 else 'NaN')   

df_soccer_full.head()  
```
